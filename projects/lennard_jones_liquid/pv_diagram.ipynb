```python
# data mining
import math
%matplotlib inline
from matplotlib import pyplot as plt
import numpy as np
import os
import glob
from itertools import cycle
```

```python
path = "out.txt"

dataset = []
with open(path, 'r') as f:
    while True:
        line = f.readline()
        if not line:
            break
        dataset.append(line.rstrip().split(";"))

dataset.sort()
dataset
```

```python
pressures = []
deviations = []
densities = [1.2, 2.4, 4.8]
temperatures = [120, 240, 300]

def func(arr, t):
    res = []
    for x in arr:
        res.append(x * (t/300))
    return res

f = plt.figure()
f.set_figwidth(10)
f.set_figheight(10)

plt.title("pV diagram")
plt.xlabel("Density x 10^(-5) [parts/[A]^3]")
plt.ylabel("Pressure x 10^(-5) [kT/[A]^-3]")

i=0
for p in pressures:
    #plt.plot(densities, p, '--', color='gray')
    plt.errorbar(densities, p, deviations[i], linestyle='None', ecolor='gray', capsize=3, capthick=3)
    plt.plot(densities, p, 'x', label=f"kT: {temperatures[i]} K")
    i+=1

for t in temperatures:
    plt.plot(densities, func(densities, t), linestyle="--")

plt.legend(loc="upper left")
plt.show()
```
